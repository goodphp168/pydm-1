### 1. 讀取 log 檔 goodeploy_log.csv 內容，計算各 function 被叫用的次數

````python
csv = 'goodeploy_log.csv'

count = {}

# count
with open(csv, encoding='utf-8') as f:
    for row in f:
        items = row.split(',')
        if len(items) == 6:
            end = items[2].find(']')
            name = items[2][2:end]
            count[name] = count.get(name, 0) + 1

# output
name_len = max(len(name) for name in count)
value_len = max(len(str(count[name])) for name in count)
for name in count:
    print(f'{name:{name_len}} {count[name]:{value_len}} times')
````

### 2. 將上述結果輸出至檔案

````python
csv = 'goodeploy_log.csv'

count = {}

# count
with open(csv, encoding='utf-8') as f:
    for row in f:
        items = row.split(',')
        if len(items) == 6:
            end = items[2].find(']')
            name = items[2][2:end]
            count[name] = count.get(name, 0) + 1

# output
output = 'output.txt'
name_len = max(len(name) for name in count)
value_len = max(len(str(count[name])) for name in count)
with open(output, 'w', encoding='utf-8') as f:
    for name in count:
        print(f'{name:{name_len}} {count[name]:{value_len}} times', file=f)
print(f'output to {output} done')
````
